```python

def explore_health_data(df, symptom_cols=None):
    """
    Explore health-related dataset:
    - Age groups
    - Sex distribution
    - Symptom frequencies
    - Comorbidities
    - Outcomes
    
    Args:
        df (pd.DataFrame): Input data
        symptom_cols (list, optional): Columns to treat as symptoms. 
                                       Defaults to ["symptoms", "cough", "fever", "difficulty_breathing"].
    Returns:
        dict: Results stored in dictionaries for later analysis
    """
    
    if symptom_cols is None:
        symptom_cols = ["symptoms", "cough", "fever", "difficulty_breathing"]

    results = {}

    # 1. AGE GROUPS
    print("\nðŸ“Š Age groups distribution:")
    if "age" in data_clean.columns:
        res = data_clean["age"].value_counts(dropna=False)
        print(res)
        results["age"] = res
    else:
        print("âš ï¸ Column 'age' not found")

    # 2. SEX DISTRIBUTION
    print("\nðŸ“Š Sex distribution (%):")
    if "sex" in data_clean.columns:
        res = (data_clean["sex"].value_counts(dropna=False, normalize=True) * 100).round(2)
        print(res)
        results["sex"] = res
    else:
        print("âš ï¸ Column 'sex' not found")

    # 3. SYMPTOMS
    print("\nðŸ“Š Symptoms frequency:")
    results["symptoms"] = {}
    for col in symptom_cols:
        if col in data_clean.columns:
            res = data_clean[col].value_counts(dropna=False)
            print(f"\n{col}:")
            print(res)
            results["symptoms"][col] = res
        else:
            print(f"âš ï¸ Column '{col}' not found")

    # 4. COMORBIDITIES
    print("\nðŸ“Š Comorbidity frequency:")
    if "comobidity" in data_clean.columns:  # keep your spelling
        res = data_clean["comobidity"].value_counts(dropna=False)
        print(res)
        results["comorbidity"] = res
    else:
        print("âš ï¸ Column 'comobidity' not found")

    # 5. OUTCOMES
    print("\nðŸ“Š Outcome status:")
    if "outcome" in data_clean.columns:
        res = data_clean["outcome"].value_counts(dropna=False)
        print(res)
        results["outcome"] = res
    else:
        print("âš ï¸ Column 'outcome' not found")

    return results
```

#### Deaths/ Recoveries over time


##### 1. Date Sample collected

```python
# Drop rows with empty strings
age_clean = age_clean[age_clean["date_sample_collected"].notna()] 
age_clean = age_clean[age_clean["date_sample_collected"].str.strip() != ""]

# Drop rows that contain "1900"
age_clean = age_clean[~age_clean["date_sample_collected"].str.contains("1900")]

# Check the result
age_clean['date_sample_collected'].info()
age_clean['date_sample_collected'].head(10)
```

#### 2.Onset Date

```python
# Drop NaN values first
age_clean = age_clean[age_clean["onset_date"].notna()]

# Convert to string
age_clean["onset_date"] = age_clean["onset_date"].astype(str)

# Drop empty strings
age_clean = age_clean[age_clean["onset_date"].str.strip() != ""]

# Drop rows containing "Asymptomatic" or "0001"
age_clean = age_clean[~age_clean["onset_date"].str.contains("Asymptomatic|0001", na=False)]

# Step 2: Convert to datetime
age_clean["onset_date"] = pd.to_datetime(age_clean["onset_date"], errors="coerce")

# Step 3: Drop rows that could not be converted (i.e., became NaT)
age_clean = age_clean[age_clean["onset_date"].notna()]

# Check results
age_clean['onset_date'].info()
age_clean['onset_date'].head(10)
```

```python
# Group by date and outcome
outcome_trends = (
    age_clean
    .groupby([age_clean["onset_date"].dt.date, "outcome"])
    .size()
    .unstack(fill_value=0)
)

# Plot
outcome_trends.plot(kind="line", figsize=(12,6))
plt.title("Deaths & Recoveries Over Time")
plt.xlabel("Date of Outcome")
plt.ylabel("Number of Cases")
plt.legend(title="Outcome")
plt.tight_layout()
plt.show()
```
